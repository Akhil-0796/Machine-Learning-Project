```python
# kernel density estimation (KDE) is a non-parametric way to estimate the probability density function of a random variable.
i = 1

fig, ax = plt.subplots(8,4,figsize=(18,30))
for column_header in column_headers:    
    plt.subplot(8,4,i)
    sns.kdeplot(fraud[column_header], bw = 0.4, label = "Fraud", shade=True, color="r", linestyle="--")
    sns.kdeplot(non_fraud[column_header], bw = 0.4, label = "Non Fraud", shade=True, color= "y", linestyle=":")
    plt.title(column_header, fontsize=12)
    i = i + 1
plt.show();
```

# SPLITTING DATA_SET INTO TEST_SET AND TRAIN_SET AND CLEANING 

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data_set['Amount_Norm']=sc.fit_transform(data_set['Amount'].values.reshape(-1,1))
```

```python
data_set
```

```python
data_set=data_set.drop(['Amount'],axis=1)
```

```python
# drop target label column
X=data_set.drop(['Class'],axis=1)
y=data_set['Class']
```

```python
# splitting data_set into train_set and test_set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

# TRAINING THE DATA_SET

```python
from sklearn.naive_bayes import GaussianNB
NB_classiifier=GaussianNB()
NB_classiifier.fit(X_train,y_train)
```

# EVALUATING THE MODEL

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
y_predict_train=NB_classiifier.predict(X_train)
y_predict_train
```

```python
cm=confusion_matrix(y_train,y_predict_train)
sns.heatmap(cm,annot=True)
```

```python
# predicting then test_set results

y_predict_test=NB_classiifier.predict(X_test)
```

```python
cm=confusion_matrix(y_test,y_predict_test)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_predict_test))
```

# IMPROVING THE MODEL

```python
X=data_set.drop(['Time','V8','V13','V15','V20','V22','V23','V24','V25','V26','V27','V28','Class'],axis=1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
NB_classiifier.fit(X_train,y_train)
y_predict=NB_classiifier.predict(X_test)
cm=confusion_matrix(y_test,y_predict)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_predict))
```
